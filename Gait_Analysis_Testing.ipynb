```python
#Converting the SensorDataTesting1.JSON and SensorDataTesting2.Json JSON data to csv data

import csv
import json


## Function to convert the SensorData.Json to SensorData.csv

def convert_json_to_csv(json_file_path, csv_file_path):

    with open(json_file_path, 'r') as json_file:
        data = json.load(json_file)


    with open(csv_file_path, 'w', newline='') as csv_file:
        writer = csv.writer(csv_file)


        headers = ['id'] + list(next(iter(data.values())).keys())
        writer.writerow(headers)


        for key, values in data.items():
            row = [key] + list(values.values())
            writer.writerow(row)
```

Converting JSON TO CSV

```python
convert_json_to_csv('SensorDataTesting1.json', 'SensorDataTesting1.csv')
convert_json_to_csv('SensorDataTesting2.json', 'SensorDataTesting2.csv')
```
